```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
file_path = r"C:\...\New Annotated Dataset.xlsx"

#Read Excel sheets
human_df = pd.read_excel(file_path, sheet_name=0)   #first sheet
gpt_df = pd.read_excel(file_path, sheet_name=1)     #second sheet

#Making sure that the column are recognised correctly
print("Human columns:", human_df.columns)
print("GPT columns:", gpt_df.columns)
```

```python
criteria = [
    "Data availability",
    "Code availability",
    "Inclusion of a ReadMe",
    "Trained model inclusion",
    "Hyperparameter description",
    "Training description",
    "Paper readability"
]

#Main function for the Cohen’s kappa calculation
kappa_results = {}

for col in criteria:
    if col not in human_df.columns:
        print(f"Missing in human_df: {col}") #Error handling
        continue
    if col not in gpt_df.columns:
        print(f"Missing in gpt_df: {col}") #Error handling
        continue

    kappa = cohen_kappa_score(human_df[col], gpt_df[col])
    kappa_results[col] = kappa
```

```python
#Overall mean kappa for all criteria
mean_kappa = sum(kappa_results.values()) / len(kappa_results)

print(f"\nOverall mean kappa: {mean_kappa:.3f}")
```
